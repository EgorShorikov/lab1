#Задача 1
Написать функцию на вход которой подается строка, состоящая из латинских букв.
Функция должна вернуть количество гласных букв (a, e, i, o, u) в этой строке. Написать тесты для кода

```python
import unittest


def get_vowels(string: str) -> dict:
    vowels = ["a", "e", "i", "o", "u"]
    return sum({k: v for (k, v) in zip(vowels, [string.count(x) for x in vowels])}.values())


class TestGetVowels(unittest.TestCase):
    def test_get_vowels(self):
        self.assertEqual(get_vowels("aaaaaiiiijo"), 10)
        self.assertEqual(get_vowels("bbb"), 0)
        self.assertEqual(get_vowels("aeiou"), 5)
        self.assertEqual(get_vowels("aaaaaa"), 6)
        self.assertEqual(get_vowels("amogus"), 3)
        self.assertEqual(get_vowels("baob"), 2)
        self.assertEqual(get_vowels("a"*100), 100)


loader = unittest.TestLoader()
suite = loader.loadTestsFromTestCase(TestGetVowels)
runner = unittest.TextTestRunner()
result = runner.run(suite)
```

# Задача 2
Написать функцию на вход, которой подается строка. Функция должна вернуть true, если
каждый символ в строке встречается только 1 раз, иначе должна вернуть false. Написать тесты для кода

```python
import unittest


def is_every_char_once(string: str) -> bool:
    char_count = {}

    for char in string:
        if char not in char_count:
            char_count[char] = 0
        char_count[char] += 1

    for count in char_count.values():
        if count != 1:
            return False
    return True


class TestIsEveryCharOnes(unittest.TestCase):

    def test_empty_string(self):
        self.assertTrue(is_every_char_once(""))

    def test_single_character(self):
        self.assertTrue(is_every_char_once("a"))
        self.assertTrue(is_every_char_once("1"))
        self.assertTrue(is_every_char_once("!"))

    def test_all_unique_chars(self):
        self.assertTrue(is_every_char_once("abcde"))
        self.assertTrue(is_every_char_once("12345"))
        self.assertTrue(is_every_char_once("aBcDe"))
        self.assertTrue(is_every_char_once("!@#$%"))

    def test_duplicate_chars(self):
        self.assertFalse(is_every_char_once("hello"))
        self.assertFalse(is_every_char_once("apple"))
        self.assertFalse(is_every_char_once("112233"))
        self.assertFalse(is_every_char_once("test"))

    def test_case_sensitivity(self):
        self.assertTrue(is_every_char_once("Aa"))
        self.assertTrue(is_every_char_once("AbCd"))
        self.assertFalse(is_every_char_once("Aaa"))

    def test_whitespace(self):
        self.assertTrue(is_every_char_once("a b"))
        self.assertFalse(is_every_char_once("  "))

    def test_special_characters(self):
        self.assertTrue(is_every_char_once("!@$"))
        self.assertFalse(is_every_char_once("!!"))
        self.assertTrue(is_every_char_once("a1!B2@f"))

    def test_long_strings(self):
        self.assertTrue(is_every_char_once("abcdefghijklmnopqrstuvwxyz"))
        self.assertFalse(is_every_char_once("abcdefghijklmnopqrstuvwxyza"))
        self.assertTrue(is_every_char_once("1234567890"))


loader = unittest.TestLoader()
suite = loader.loadTestsFromTestCase(TestIsEveryCharOnes)
runner = unittest.TextTestRunner()
result = runner.run(suite)
```

# Задача 3
Написать функцию, которая принимает положительное число и возвращает количество
бит равных 1 в этом числе. Написать тесты для кода

```python
import unittest


def count_one_bits(num: int) -> int:
    binary_num = bin(num)[2:]
    return binary_num.count("1")


class TestCountOneBits(unittest.TestCase):
    def test_zero(self):
        self.assertEqual(count_one_bits(0), 0)

    def test_two_powers(self):
        for power in range(10):
            num = 2 ** power
            self.assertEqual(count_one_bits(num), 1)
    
    def test_large_numbers(self):
        self.assertEqual(count_one_bits(999), 8)
        self.assertEqual(count_one_bits(789), 5)

    def test_every_char_is_one(self):
        for power in range(2, 10):
            num = 2 ** power - 1
            self.assertEqual(count_one_bits(num), power)


loader = unittest.TestLoader()
suite = loader.loadTestsFromTestCase(TestCountOneBits)
runner = unittest.TextTestRunner()
result = runner.run(suite)
```

# Задача 4
Написать функцию, которая принимает положительное число. Функция должна вернуть то,
сколько раз необходимо перемножать цифры числа или результат перемножения, чтобы
получилось число состоящее из одной цифры.
Например, для входного числа:
· 39 функция должна вернуть 3, так как 3*9=27 => 2*7=14 => 1*4=4
· 4 функция должна вернуть 0, так как число уже состоит из одной цифры
· 999 функция должна вернуть 4, так как 9*9*9=729 => 7*2*9=126 => 1*2*6=12 =>
1*2=2. Написать тесты для кода

```python
import unittest


def multiply_chars(num: int):
    result = 1
    for char in str(num):
        result *= int(char)
    return result


def magic(n: int) -> int:
    step = 0
    if n < 10 or "0" in str(n):
        return step
    while n > 10:
        n = multiply_chars(n)
        step += 1
    return step


class TestMagic(unittest.TestCase):
    def test_example(self):
        self.assertEqual(magic(39), 3)
        self.assertEqual(magic(999), 4)

    def test_one_digit(self):
        self.assertEqual(magic(1), 0)
        self.assertEqual(magic(3), 0)
        self.assertEqual(magic(4), 0)
        self.assertEqual(magic(7), 0)
        self.assertEqual(magic(9), 0)

    def test_numbers_with_zero(self):
        self.assertEqual(magic(10), 0)
        self.assertEqual(magic(100), 0)
        self.assertEqual(magic(101), 0)
        self.assertEqual(magic(123405678), 0)

    def test_two_digits(self):
        self.assertEqual(magic(24), 1)
        self.assertEqual(magic(29), 2)
        self.assertEqual(magic(89), 3)
        self.assertEqual(magic(99), 2)

    def test_three_digits(self):
        self.assertEqual(magic(256), 2)
        self.assertEqual(magic(322), 2)
        self.assertEqual(magic(789), 2)


loader = unittest.TestLoader()
suite = loader.loadTestsFromTestCase(TestMagic)
runner = unittest.TextTestRunner()
result = runner.run(suite)
```

# Задача 5
Написать функцию, которая принимает два целочисленных вектора одинаковой длины и
возвращает среднеквадратическое отклонение двух векторов. Написать тесты для кода

```python
import unittest


from math import sqrt


def mse(pred, true):
    dispersion = sum([(i - j)**2 for i, j in zip(pred, true)])
    return round(sqrt(dispersion), 2)


class TestMSE(unittest.TestCase):
    def test_same_vectors(self):
        self.assertEqual(mse([1, 2], [1, 2]), 0.0)
        self.assertEqual(mse([1, 2, 3], [1, 2, 3]), 0.0)
        self.assertEqual(mse([1, 2, 3, 4], [1, 2, 3, 4]), 0.0)
        
    def test_zero_length_vectors(self):
        self.assertEqual(mse([], []), 0.0)

    def test_inverted_vectors(self):
        self.assertEqual(mse([2], [3]), 1.0)
        self.assertEqual(mse([3], [2]), 1.0)
        self.assertEqual(mse([-1], [2]), 3.0)
        self.assertEqual(mse([2], [-1]), 3.0)

    def test_three_dimension_vectors(self):
        self.assertEqual(mse([2, 5, 9], [3, 2, 2]), 7.68)
        self.assertEqual(mse([-19, 5, 39], [20, 1, 30]), 40.22)
        self.assertEqual(mse([7, 2, -3], [2, -13, 10]), 20.47)


loader = unittest.TestLoader()
suite = loader.loadTestsFromTestCase(TestMSE)
runner = unittest.TextTestRunner()
result = runner.run(suite)
```

# Задача 6
Написать функцию, принимающая целое положительное число. Функция должна вернуть
строку вида “(n1**p1)(n2**p2)…(nk**pk)” представляющая разложение числа на простые
множители (если pi == 1, то выводить только ni).
Например, для числа 86240 функция должна вернуть “(2**5)(5)(7**2)(11)”. Написать тесты для кода

```python
import unittest


def get_primes(n):
    primes = []
    m = [False, False] + [True] * (n - 1)
    p = 2
    while p**2 < n:
        for div in range(p**2, n + 1, p):
            if div != p:
                m[div] = False
        for div in range(p + 1, n + 1):
            if div == n:
                return div

            if m[div] == True:
                p = div
                break

    for is_prime in range(len(m)):
        if m[is_prime] == True:
            primes.append(is_prime)

    return primes


def prime_factorization(n):
    primes = get_primes(n)
    div_primes = {}
    while n != 1:
        for prime in primes:
            if n % prime == 0:
                if prime not in div_primes:
                    div_primes[prime] = 0
                div_primes[prime] += 1
                n //= prime
    return div_primes


def print_factorization(div_primes):
    m = []
    for prime, prime_count in div_primes.items():
        if prime_count != 1:
            m.append(f"({prime}**{prime_count})")
        else:
            m.append(f"({prime})")
    return "".join(m)


class TestPrintFactorization(unittest.TestCase):
    def test_example(self):
        div_primes = prime_factorization(86240)
        self.assertEqual(print_factorization(div_primes), "(2**5)(5)(7**2)(11)")
    
    def test_primes(self):
        div_primes = prime_factorization(2)
        self.assertEqual(print_factorization(div_primes), "(2)")
        
        div_primes = prime_factorization(3)
        self.assertEqual(print_factorization(div_primes), "(3)")

        div_primes = prime_factorization(7)
        self.assertEqual(print_factorization(div_primes), "(7)")

        div_primes = prime_factorization(29)
        self.assertEqual(print_factorization(div_primes), "(29)")

    def test_big_nums(self):
        div_primes = prime_factorization(123456)
        self.assertEqual(print_factorization(div_primes), "(2**6)(3)(643)")

        div_primes = prime_factorization(1234567)
        self.assertEqual(print_factorization(div_primes), "(127)(9721)")

        div_primes = prime_factorization(12345678)
        self.assertEqual(print_factorization(div_primes), "(2)(3**2)(47)(14593)")
    
    def test_powers_of_primes(self):
        tests = {
            4: "(2**2)",
            8: "(2**3)", 
            9: "(3**2)",
            16: "(2**4)",
            25: "(5**2)",
            27: "(3**3)",
            32: "(2**5)",
            49: "(7**2)",
        }
        for n, expected in tests.items():
            div_primes = prime_factorization(n)
            self.assertEqual(print_factorization(div_primes), expected)

    def test_other(self):
        tests = {
            120: "(2**3)(3)(5)",
            322: "(2)(7)(23)", 
            666: "(2)(3**2)(37)",
            1024: "(2**10)",
            2048: "(2**11)"
        }
        for n, expected in tests.items():
            div_primes = prime_factorization(n)
            self.assertEqual(print_factorization(div_primes), expected)


loader = unittest.TestLoader()
suite = loader.loadTestsFromTestCase(TestPrintFactorization)
runner = unittest.TextTestRunner()
result = runner.run(suite)
```

# Задача 7
Написать функцию, принимающая целое число n, задающее количество кубиков. Функция
должна определить, можно ли из данного кол-ва кубиков построить пирамиду, то есть
можно ли представить число n как 1^2+2^2+3^2+…+k^2. Если можно, то функция должна
вернуть k, иначе строку “It is impossible”. Написать тесты для кода

```python
import unittest


def pyramid(number):
    if number == 0:
        return "It is impossible"
    k = 0
    while number > 0:
        k += 1
        number -= k ** 2
    return k if number == 0 else "It is impossible"


class TestPyramid(unittest.TestCase):
    def test_possible(self):
        self.assertEqual(pyramid(1), 1)
        self.assertEqual(pyramid(5), 2)
        self.assertEqual(pyramid(14), 3)
        self.assertEqual(pyramid(30), 4)
        self.assertEqual(pyramid(91), 6)

    def test_impossible(self):
        self.assertEqual(pyramid(0), "It is impossible")
        self.assertEqual(pyramid(20), "It is impossible")
        self.assertEqual(pyramid(200), "It is impossible")
        self.assertEqual(pyramid(314), "It is impossible")
        self.assertEqual(pyramid(777), "It is impossible")
        self.assertEqual(pyramid(1000), "It is impossible")


loader = unittest.TestLoader()
suite = loader.loadTestsFromTestCase(TestPyramid)
runner = unittest.TextTestRunner()
result = runner.run(suite)
```

# Задача 8
Функция принимает на вход положительное число и определяет является ли оно сбалансированным, т.е. сумма цифр до средних равна сумме цифр после. Средними в случае нечетного числа цифр считать одну цифру, в случае четного - две средних. Написать тесты для кода

```python
import unittest


def is_num_balanced(num: int):
    num_as_string = str(num)
    half_length = len(num_as_string) // 2
    if len(num_as_string) % 2 != 0:
        return sum([int(x) for x in num_as_string[:half_length]]) == sum([int(x) for x in num_as_string[half_length + 1:]])
    
    return sum([int(x) for x in num_as_string[:half_length]]) == sum([int(x) for x in num_as_string[half_length:]])
    

class TestIsNumBalanced(unittest.TestCase):
    def test_even_true(self):
        self.assertTrue(is_num_balanced(11))
        self.assertTrue(is_num_balanced(22))
        self.assertTrue(is_num_balanced(1221))
        self.assertTrue(is_num_balanced(3131))
        self.assertTrue(is_num_balanced(123510))
        self.assertTrue(is_num_balanced(6693))

    def test_even_false(self):
        self.assertFalse(is_num_balanced(1234))
        self.assertFalse(is_num_balanced(6677))
        self.assertFalse(is_num_balanced(1121))
        self.assertFalse(is_num_balanced(1011))
        self.assertFalse(is_num_balanced(777778))
    
    def test_odd_true(self):
        self.assertTrue(is_num_balanced(101))
        self.assertTrue(is_num_balanced(404))
        self.assertTrue(is_num_balanced(13031))
        self.assertTrue(is_num_balanced(29283))
        self.assertTrue(is_num_balanced(11111))

    def test_odd_false(self):
        self.assertFalse(is_num_balanced(122))
        self.assertFalse(is_num_balanced(322))
        self.assertFalse(is_num_balanced(10002))
        self.assertFalse(is_num_balanced(726))
        self.assertFalse(is_num_balanced(29110))


loader = unittest.TestLoader()
suite = loader.loadTestsFromTestCase(TestIsNumBalanced)
runner = unittest.TextTestRunner()
result = runner.run(suite)
```
